```python
import requests
from datetime import datetime, timedelta
import pytz

# Set your timezone and calculate 1 hour before now
timezone = pytz.timezone("Asia/Kolkata")
one_hour_ago = datetime.now(timezone) - timedelta(hours=1)
start_time_iso = one_hour_ago.isoformat()

# Query parameters
query = "bangalore"
url = "https://twitterapi.io/api/v2/search"

headers = {
    "Authorization": "Bearer 8934b12ef9194b7ca525f769c112f781"
}

params = {
    "q": query,
    "since": start_time_iso,
    "count": 100  # Maximum per call
}

response = requests.get(url, headers=headers, params=params)

if response.status_code == 200:
    tweets = response.json()
    for tweet in tweets.get("data", []):
        print(f"@{tweet['user']['screen_name']}: {tweet['full_text']}\n")
else:
    print("Error:", response.status_code, response.text)
```
